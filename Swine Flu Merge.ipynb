```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
IXIC['Date'] = pd.to_datetime(IXIC['Date'])
```

```python
IXIC.dtypes
```

```python
Swine = pd.read_csv('Swine_Flu.csv')
Swine
```

```python
Dates = pd.to_datetime(Swine.year.astype(str), format='%Y') + pd.to_timedelta(Swine.week.mul(7).astype(str) + ' days')
```

```python
Dates
```

```python
Swine['Date'] = Dates
Swine
```

```python
Swine_DJI = pd.merge(Swine, DJI, left_on = 'Date', right_on = 'Date')
Swine_DJI
```

```python
Swine_DJI.to_csv('Swine_DJI.csv')
```

```python
print('The NA values present in Swine_DJI:\n', Swine_DJI.isnull().sum(), '\n')
print('The data types in Swine_DJI:\n', Swine_DJI.dtypes, '\n')
print('The data shape of Swine_DJI:\n', Swine_DJI.shape, '\n')
```

```python
Swine_GSPC = pd.merge(Swine, GSPC, left_on = 'Date', right_on = 'Date')
Swine_GSPC
```

```python
Swine_GSPC.to_csv('Swine_GSPC.csv')
```

```python
print('The NA values present in Swine_GSPC:\n', Swine_GSPC.isnull().sum(), '\n')
print('The data types in Swine_GSPC:\n', Swine_GSPC.dtypes, '\n')
print('The data shape of Swine_GSPC:\n', Swine_GSPC.shape, '\n')
```
