```python
import pandas as pd
import numpy as np
```

```python
DJI = pd.read_csv('^DJI.csv')
DJI
```

```python
GSPC = pd.read_csv('^GSPC.csv')
GSPC
```

```python
IXIC = pd.read_csv('^IXIC.csv')
IXIC
```

```python
Ebola = pd.read_csv('Ebola.csv')
Ebola
```

```python
Ebola = Ebola.rename(columns = {'date':'Date'})
Ebola
```

```python
Ebola = Ebola.groupby('Date').sum()
Ebola = Ebola.reset_index()
Ebola
```

```python
Ebola_DJI = pd.merge(Ebola, DJI, left_on = 'Date', right_on = 'Date')
Ebola_DJI
```

```python
Ebola_DJI['Date'] = pd.to_datetime(Ebola_DJI['Date'])
```

```python
Ebola_DJI.to_csv('Ebola_DJI.csv')
```

```python
print('The NA values present in Ebola_DJI:\n', Ebola_DJI.isnull().sum(), '\n')
print('The data types in Ebola_DJI:\n', Ebola_DJI.dtypes, '\n')
print('The data shape of Ebola_DJI:\n', Ebola_DJI.shape, '\n')
```

```python
Ebola_GSPC = pd.merge(Ebola, GSPC, left_on = 'Date', right_on = 'Date')
Ebola_GSPC
```

```python
Ebola_GSPC['Date'] = pd.to_datetime(Ebola_GSPC['Date'])
```

```python
Ebola_GSPC.to_csv('Ebola_GSPC.csv')
```

```python
print('The NA values present in Ebola_GSPC:\n', Ebola_GSPC.isnull().sum(), '\n')
print('The data types in Ebola_GSPC:\n', Ebola_GSPC.dtypes, '\n')
print('The data shape of Ebola_GSPC:\n', Ebola_GSPC.shape, '\n')
```

```python
Ebola_IXIC = pd.merge(Ebola, IXIC, left_on = 'Date', right_on = 'Date')
Ebola_IXIC
```

```python
Ebola_IXIC['Date'] = pd.to_datetime(Ebola_IXIC['Date'])
```

```python
Ebola_IXIC.to_csv('Ebola_IXIC.csv')
```
