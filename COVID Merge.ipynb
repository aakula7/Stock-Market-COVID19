```python
import pandas as pd
import numpy as np
```

```python
DJI = pd.read_csv('^DJI.csv')
DJI
```

```python
GSPC = pd.read_csv('^GSPC.csv')
GSPC
```

```python
IXIC = pd.read_csv('^IXIC.csv')
IXIC
```

```python
COVID = pd.read_csv('COVID19.csv')
COVID
```

```python
COVID = COVID.rename(columns = {'date':'Date'})
COVID
```

```python
COVID = COVID.groupby('Date').sum()
COVID = COVID.reset_index()
COVID
```

```python
COVID_DJI = pd.merge(COVID, DJI, left_on = 'Date', right_on = 'Date')
COVID_DJI
```

```python
COVID_DJI['Date'] = pd.to_datetime(COVID_DJI['Date'])
```

```python
COVID_DJI.to_csv('COVID_DJI.csv')
```

```python
print('The NA values present in COVID_DJI:\n', COVID_DJI.isnull().sum(), '\n')
print('The data types in COVID_DJI:\n', COVID_DJI.dtypes, '\n')
print('The data shape of COVID_DJI:\n', COVID_DJI.shape, '\n')
```

```python
COVID_GSPC = pd.merge(COVID, GSPC, left_on = 'Date', right_on = 'Date')
COVID_GSPC
```

```python
COVID_GSPC['Date'] = pd.to_datetime(COVID_GSPC['Date'])
```

```python
COVID_GSPC.to_csv('COVID_GSPC.csv')
```

```python
print('The NA values present in COVID_GSPC:\n', COVID_GSPC.isnull().sum(), '\n')
print('The data types in COVID_GSPC:\n', COVID_GSPC.dtypes, '\n')
print('The data shape of COVID_GSPC:\n', COVID_GSPC.shape, '\n')
```

```python
COVID_IXIC = pd.merge(COVID, IXIC, left_on = 'Date', right_on = 'Date')
COVID_IXIC
```

```python
COVID_IXIC['Date'] = pd.to_datetime(COVID_IXIC['Date'])
```

```python
COVID_IXIC.to_csv('COVID_IXIC.csv')
```
